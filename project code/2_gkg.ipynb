```python
df2=pd.read_csv("sites.csv")
list_sites=df2.sites.to_list()
list_source=df.SOURCES.to_list()
length_sites=len(list_sites)
length_source=len(list_source)


new_index=[]
for i in range(length_source):
    for j in range(length_sites):
        if list_sites[j] in list_source[i]:
            # print(i)
            new_index.append(i)
            break
print(len(new_index))
```

```python
print(df.shape)
df=df.iloc[new_index]

print(df.shape)
df.head()
```
